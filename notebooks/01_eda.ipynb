```python
sns.countplot(x = 'y', data = df)
plt.title("Term Deposit Subscription Distribution")
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.histplot(df['age'],bins = 30, kde=True)
plt.title("Age Distribution of Clients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='job', data=df, order=df['job'].value_counts().index)
plt.title("Job Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['balance'], bins=50, kde=True)
plt.title("Average Yearly Balance Distribution")
plt.show()
```

```python
sns.countplot(x='marital', data=df)
plt.title("Marital Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='job', hue='y', data=df)
plt.title("Job vs Term Deposit Subscription")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='y', y='duration', data=df)
plt.title("Call Duration vs Subscription")
plt.show()
```
